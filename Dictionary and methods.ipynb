### Dictionary:
                Dictionary is ordered collection key-value(i.e. item) pairs.It is mutable and defines with '{}'
                Dictionary formate:
                                {Key:value}
                Characteristics:
                        1.Dictionary is ordered collection
                        2.Dictionary is mutable
                        3.Dictionary does not support indexing .
                        4.Dictionary keys are uniques.
                        5.Dictionary values allows duplicates.
                        6.List can't be used as dictionary key,because unhashable.
                        7.Dictionary values cn access by using keys

                example:
                    details={'name':"rakesh",'course':"python",'batch':23}
                    details['name']

                    
                        Hashing it storing mechanism

```python
#Ordered collection 

details={"name":"srinu","course":"python","batch":22}
print(details)
```

```python
# Dictionary values access by using key
#Updating dictionary values
#Adding new items to dictionary

details={"name":"srinu","course":"python","batch":22,"age":45}
print(details["course"])  
#print(details["address"])# if it is not there in the list it should be return "key error"
#updating dictionary values
details["age"] = details["age"]+1
print(details["age"])
#adding  new items
details["locations"]="remidicherla"
print(details)
```

```python

details={"name":"srinu","course":"python","batch":22,"age":45,"age":55,"location":"chennai"}
print(details)
```

```python
#List can't be used as dictionary key
#dictionary key are unique
details={"name":"srinu","course":"python",[1,2,3,4] : 22,"batch":22,"age":45,"marks":34,}
print(details)
```


### Dictionary Methods
        1.get(Key,default_values:optional): It returns the values if key is exists in dictionary,otherwise it returns defaul value.



```python
#dictionary key and uniue

details={"name":"srinu","course":"python","batch":22,"age":45,"marks":234}
print(details)
print(details["place"]) #this one raise KeyError
print(details["name"])
print(details.get("course"))
print(details.get("marks")) 
print(details.get("place")) #get method by default it return none if key not exits
print(details.get("place","place not available"))  #It return default values because 
```

### update(new_dictionary):
            This method updates the values if new_dicttionary keys already exits in dictionary,otherwise it adds new items to dictionary

```python
details={"name":"srinu","course":"python","batch":22,"age":45}
details.update({"batch":34,"course":"java","location":"hyd"})
print(details)
details["location"]="chennai"
print(details)
```

### 
    1.Keys(): This method returns the list dicitionary keys
    2.values(): This method returns the list dicitionary values
    3.items():This method return the list of tuple of key-values pairs

```python
details={'name': 'srinu', 'course': 'java', 'batch': 34, 'age': 45, 'location': 'hyd'}
keys=details.keys()
print(details.keys())
print(type(keys))
print(keys)
print(details.values())
print(details.items())
```

### ITEM REMOVEING ELEMENTS
        1.popitem():It return and removes the last item as tuple
        2.pop(key):This method return the specified key item and return the specified key's values,otherwise 'keyError'

```python
details={'name': 'srinu', 'course': 'java', 'batch': 34, 'age': 45, 'location': 'hyd'}
item=details.popitem()
print(item)
print(details)
val=details.pop("name")
print(val)
print(details)
val1=details.pop("age")
print(val1)
print(details)
```
