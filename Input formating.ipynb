### Reading list of integer from user
            1.input()----string
            2.input().split-----list of string
    Map(): 
        This function applies funtionally on each element in itearable .
    Syntax:
        map(fun,iterable)
                    ^
                    |
                map(int,['1','2','3','4'])-->map object

                map(int,input().split())----map object

               1. list(map(int,input().split())----list of integer


```python
list=input().split()
print(list)
```

```python
lst = list(map(int, input().split()))
print(lst)
```

```python
#reading tuple of integer  from user

lst=tuple(map(int, input().split()))
print(lst)
```

```python
#reading set of integer  from user
lst=set(map(int,input().split()))
print(lst)
```
