## Lists
    List is a ordered collection different data type elements.Lists are mutable and defines with '[]'.
        Characteristics:
           - Lists are ordered collection 
           - List are mutable
           - List allows indexing and slicing
           - List contains duplicate elements
           - List is hetrogeneous datatype(It allows different data types elements) (Homegenous menas it allows single data type elements)
           - List is unhashable datatype

                example:
                    [1,3,5,6,7,'hello',1,2,3]

```python
#Define list
lst=[1,2,3,4,"hii codegnan","hyd",1,4,3]
print(lst)
```

```python
#indexing operations
list1=[10,20,30,"hello",45]
print(list1[2])
list1[2] = list1[4]
print(list1[2])
print(list1[-2])
```

```python
name=["rakesh","veltech","srm","codegnan"]
print(name[2])
```

```python
#Slicing

list=[10,20,30,40,50,60,70]
print(list[:5:-1])
print(list[:500])
print(list[::-1]) #Reverse list
```

```python
# Empty list creation


l1 = []
l2 = list()
print(type(l1),type(l2))
```

```python
#List operations
    #1.List concation (using + operator)
    #2.List Repetition (Using * operator)
l1=[1,2,3,4,5]
l2=[2,3,4,5,6,7]
print(l1+l2)
print(l1 * 2)
print([0] * 5)
```

## List Methods
    Item Adding Methods
         1.append(item):It adds item at end of list
         2.extend(iterable):It adds the all iterator element at end of list
         3.insert(pos, item):It adds item  at specified position
        

```python
list=[10,20,30,40,50]
list.append(200)
print(list)
list.extend([1,2,3,4,5,6,7])
print(list)
list.append([120,150])
print(list)
list.insert(2,2000) ##index position and then remaining elements are move right shift.
print(list)
```

### Item Removing Method
        - remove(item):It removes the first occurance item from the list if it is exists in list otherwise 'ValueError'
        - pop(pos):It remove and return the specificed position item if position is passed , otherwise it remove and returns the last  item from list
        - clear():It remove all elements from list

```python
list=[10,20,30,40,20,40,40]
print(list.remove(40))  #index is not specified it will show value error
print(list)
list=[12,33,4,4,6,7]
print(list.remove(4))
print(list)
```

```python
lst=[10,20,30,40,50,30,20]
val=lst.pop() #remove the last elements of the list
print(val)
print(lst)
var2=lst.pop(2) #it remove the oth position
print(var2)
print(lst)
lst.clear() # It clear the list
print(lst)
```

```python
#index and count()
   # - index(item): If item is present in list it return first occurance the index otherwise ValueError
   # - count(item):It return the count of total occurances of item

lst=[10,20,30,40]
ind=lst.index(30)
print(ind)
print(lst.count(20))  
print(lst.count(100)) # It return 0 if values not in the list
```

```python
## Sorting
   # - sort(reverse = False) : By default it sorts the origina list in in into Ascending order.
   # -                             if reverser is True it sorts in Descending order

list=[1,2,3,4,7,4,5,3,-34,-56,-78]
list.sort()  # list should be in ascending order
print(list)
list.sort(reverse =True) # list should be in descending order
print(list)
```

```python
name=["rakesh","veltech","codegnan"]
name.sort()
print(name)
name.sort(reverse =True)
print(name)
```
