### Sets:
        Sets is unordered collection of different datatype unique elements.sets are mutable and defines with '{}'
    Characteristics:
        1.set is unordered collection
        2.set elements are unique.
        3.set does not support indexing
        4.set are mutable
        5.set allows hashable datatype as elements
        6.set elements can be used as dictionary keys.
