## String Method
- Case Conversion method
    - Upper():It converts entire string into upper case (Capital letters)
    - lower():It converts entire string into Lower case (Small letters)
    - title():It converts the first letter of every word into upper and remain in lower
    - swapcase():It converts upper case letter into lower case  and lower to upper
    - capitalize():It converts first character into upper case remaining into lower case
      

```python
string="I am studing python full stack"
print(string.upper())
print(string)
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

# Remove White space or pattern at end of string
    - strip(): By default it remove the white space at both ends of string
    - lstrip(): It removes the white space at left side of string
    - rstrip(): It removes the white space at right side of string
    

```python
s="            RAKESH       "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
r="            RAKESH       "
print(r)
print(r.strip())
print(r.lstrip())
print(r.rstrip())
```

```python
s="            RAKESH       "
print("@")
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
#Split(): It splite the entire string into list of substring based on specified separate
#by default it splits at space 

string="I am studing python full  stack at codegnan"
print(string.split())
print(string.split("am"))
print(string.split('p'))
```

```python
#'sep'.join(iterable):It joins the all iterable elements with separator
words=['I', 'am', 'studing', 'python', 'full', 'stack', 'at', 'codegnan']
print("".join(words))
print(" ".join(words))
print("-".join(words))
```

# Substring Index Finding
   - Indexing(Substring):if substring is exists in string, it returns the first occurence of substring otherwise  'ValueError'
   - Find(substring):if substring is exists in string, it returns the first occurence of substring otherwise  -1

```python
string="I am studing python full  stack at codegnan"
print(string.index("full"))
```

```python
string="I am studing python full  stack at codegnan full"
print(string.find("full"))
print(string.find("r"))
print(string.find("k"))
print(string.find("v"))
print(string.find("full",20)) #it start checking from 10th position
```

```python
#Count(Substring): It returns the total non-overlapping occucences of substring
s="ABCDBCDabcdabcd"
print(s.count("B"))
upper_string=s.upper()
print(upper_string)
print(upper_string.count("B"))
```

```python
#replace(old_substring,new_substring): It replace the old sub-string with new sub string
string="The sky is blue and the ocean is blue"
print(string.replace("blue","red"))
print(string.replace('e','E'))
print(string.replace('e','E', 3)) #IT REPLACE FIRST 2 OCCURANCE
```

###String Checking Method
    **String checking methoids return the boolean values as result**
        isupper():It return true,if entire string is in upper case
        islower():It return true,if entire string is in lower case
        isalpha():It return true,if entire string contains alphabets
        isdigit():It return true,if entire string contains digit
        isalnum():It return true,if entire string contains either alphabets or digits
        startswith(substring):It returns True if,string starts with substring
        endswith(substring):It returns True if,string ends with substring

```python
print("code".isupper())
print("codes".islower())
print("1234".isdigit())
print("code123".isalpha())
print("hello".isupper())
```
