## TUPLE:
        Tuple is ordered collection of different data type elemnts .tuple is immutable and define with '()'
            -Characteristics
                - Tuple is ordered collection
                - Tuple is immutable
                - Tuple allows indexing and slicing
                - Tuple  is hetrogenous datatype
                - tuple allows duplicate item

        EXAMPLE:
            t=(1,2,3,4,5,"hellow",2,3,4)

## Difference between list and tuple
        #Tuple:
            1.Tuple immutable
            2.Tuple represent with '()'
            3.Less memory space
            4.operation speed more
        #LIST:
            1.List is Mutable
            2.List represent with []
            3.List takes more memory space compare to tuple
