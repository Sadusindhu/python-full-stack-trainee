#### Type conversion:
                    Converting one data type format into another datatype formate.
                    This is called as type conversion

            They are 2 types:
                        1.implicit type conversion
                        2.Explicit type conversion(type casting)

            1.Implicit type conversion:
                                        The python interpreter automatically converts the datatype because ,it prevents the data loss.
                                        1.Implicit type conversion can happends in arithematic operations

                                                Example:
                                                        3+5.8=8.8
            2.Explicit type conversion(type casting):
                                                    Here type conversion happends by using python built in datatype function
                                         example:   a=2.3
                                                    b=int(a)
                                                    print(b)

            Python built in datatype function:
                        1.interger-int()
                        2.float val-complex()
                        3.boolean -bool()
                        4.string-str()
                        5.list-list()
                        6.tuple-tuple()
                        7.set-set()
                        8.dictionary-dict()

            Type conversion:
                               1. integer-float,string,complex,boolean
                               2. float-int,bool,str
                               3. Bool-int,

```python
#String to int,float,bool,list,tuple,set
#if string values is number

s="123456789"
print(type(s))
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#String to int,float,bool,list,tuple,set
#if string values is float

s="123456789.150"
print(type(s))
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#list 


my_list=["r","a","k","e","s","h"]
my_str=''.join(my_list)
my_lst=[1,2,3,4,5]
my_set=set(my_lst)
my_tuple=tuple(my_lst)
print(my_str)
print(my_set)
print(my_tuple)
```

```python
#tuple
tup=('p','y','t','h','o','n')
s=''.join(tup)
print(s)
tup=[(10,20,30,40)]
lst=list(tup)
print(lst)
st=set(tup)
print(st)
print(list(tup))
```

```python
#dictionary
a={'a':10,'b':20,'c':30}
lst=list(a)
st=set(a)
print(st)
tup=tuple(a)
print(tup)
print(lst)
s=str(a)
print(s)
```
