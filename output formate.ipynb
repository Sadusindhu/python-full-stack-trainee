```python
#1.string formating using comma
name="rakesh"
age=22
print("my name is",name,"and i am",age," years old")
```

```python
#2.Access specifiers
#String - %s
#integer - %d
#float- %f

name="rakesh"
age=22
print("my name is %s and i a %d year old"%(name,age))
```

```python
#3.dot formate
name="rakesh"
age=22
print("my name is {} and i a {} year old".format(name,age))
print("my name is {} and i a {} year old".format(age,name))
```

```python
#4.f-formate---->
name="rakesh"
age=22
print(f"my name is {name} and i am {age} year old")
print(f"my name is {age} and i am {name} year old")
```
